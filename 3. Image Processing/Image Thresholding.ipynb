# Image Thresholding

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Blue Green Red channels
img = cv2.imread("../imgs/rainbow.jpg")
```

```python
plt.imshow(img)
```

```python
# Read as grayscale image
img = cv2.imread("../imgs/rainbow.jpg", 0)
```

```python
plt.imshow(img, cmap="gray")
```

```python
img.max()
```

```python
ret,thresh1 = cv2.threshold(img, 129, 255, cv2.THRESH_BINARY_INV)
```

```python
ret
```

```python
plt.imshow(thresh1)
```

```python
plt.imshow(thresh1, cmap="gray")
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
```

```python
ret
```

```python
plt.imshow(thresh1,cmap="gray")
```

```python
plt.imshow(img, cmap="gray")
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
```

```python
plt.imshow(thresh1, cmap="gray")
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC_TO)
```
