# Blurring and Smoothing

Smoothing an image can help get rid of noise, or help an application
focus on general details

often blurring or smoothing is combined with edge detection

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
def load_img():
    img = cv2.imread("../imgs/bricks.jpg")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
display_img(img)
```

```python
gamma = 1
result = np.power(img, gamma)
display_img(result)
```
