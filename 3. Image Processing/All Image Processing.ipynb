# Image Processing Assessment

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

<b> Open and display the giraffe.jpg </b>

```python
giraffe_bgr = cv2.imread("../imgs/giraffes.jpg")
giraffe_rgb = cv2.cvtColor(giraffe_bgr, cv2.COLOR_BGR2RGB)
```

```python
display_img(giraffe_bgr)
```

```python
display_img(giraffe_rgb)
```

<b>Apply a binary threshold onto the giraffe</b>

```python
ret,binary_giraffe = cv2.threshold(giraffe_bgr, 127, 255, cv2.THRESH_BINARY)
```

```python
display_img(binary_giraffe, cmap="gray")
```

<b> Open the giraffe.jpg and convert its colorspace to HSV and display the image</b>

```python
giraffe = cv2.imread("../imgs/giraffes.jpg")
```

```python
hsv_giraffe = cv2.cvtColor(giraffe, cv2.COLOR_BGR2HSV)
```

```python
display_img(hsv_giraffe)
```
