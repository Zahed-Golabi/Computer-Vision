# Image Processing Assessment

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

<b> Open and display the giraffe.jpg </b>

```python
giraffe_bgr = cv2.imread("../imgs/giraffes.jpg")
giraffe_rgb = cv2.cvtColor(giraffe_bgr, cv2.COLOR_BGR2RGB)
```

```python
display_img(giraffe_bgr)
```

```python
display_img(giraffe_rgb)
```

<b>Apply a binary threshold onto the giraffe</b>

```python
ret,binary_giraffe = cv2.threshold(giraffe_rgb, 127, 255, cv2.THRESH_BINARY)
```

```python
display_img(binary_giraffe, cmap="gray")
```

```python
giraffe_bgr = cv2.imread("../imgs/giraffes.jpg",0)
ret,binary_giraffe = cv2.threshold(giraffe_bgr, 127, 255, cv2.THRESH_BINARY)
```

<b> Open the giraffe.jpg and convert its colorspace to HSV and display the image</b>

```python
giraffe = cv2.imread("../imgs/giraffes.jpg")
```

```python
hsv_giraffe = cv2.cvtColor(giraffe, cv2.COLOR_BGR2HSV)
```

```python
display_img(hsv_giraffe)
```

<b> Create a low pass filter with a 4*4 kernel filled with values of 1/10 and then use a 2-D Convolution to blur the giraffe image (display in normal RGB) </b>

```python
kernel = np.ones(shape=(4,4), dtype=np.float32) / 10
```

```python
kernel
```

```python
blured_giraffe = cv2.filter2D(giraffe,-1,kernel)
```

```python
blured_giraffe_rgb = cv2.cvtColor(blured_giraffe, cv2.COLOR_BGR2RGB)
```

```python
display_img(blured_giraffe_rgb)
```

<b>Create a Horizontal Sobel Filter (sobelx) with a kernel size of 5 to the grayscale version of the giraffe image and then display the resulting gradient filtered version of the image</b>

```python
grayscale_giraffe = cv2.imread("../imgs/giraffes.jpg",0)
```

```python
display_img(grayscale_giraffe, cmap="gray")
```

```python
sobelx = cv2.Sobel(grayscale_giraffe, cv2.CV_64F, 1, 0, 5)
```

```python
display_img(sobelx, cmap="gray")
```

<b>Plot the color histograms for the RED,BLUE, and GREEN channel of the giraffe image</b>

```python
img = cv2.imread("../imgs/giraffes.jpg")
```

```python
channels = ["blue","green","red"]

for index,ch in enumerate(channels):
    histr = cv2.calcHist([img],[index],None,[256],[0,256])
    plt.plot(histr, color=ch, label=ch + " channel")
plt.title("MY HISTOGRAM")
plt.legend()
```
