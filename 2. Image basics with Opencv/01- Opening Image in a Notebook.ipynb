# Opening Images in a Notebook

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("../imgs/flower1.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# MATPLOTLIB  ---> RGB RED GREEN BLUE
# OPENCV      ---> BGR BLUE GREEN RED
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread("../imgs/dog.png", cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
img_gray.min()
```

```python
img_gray.max()
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
plt.imshow(fix_img)
```

```python
# Resizing images
fix_img.shape
```

```python
resized_img = cv2.resize(fix_img, (1000, 600))
```

```python
plt.imshow(resized_img)
```

```python
resized_img.shape
```

```python
# Resize with Aspect Ratio
w_ratio = 0.5
h_ratio = 0.5
```

```python
ar_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(ar_img)
```

```python
ar_img.shape
```

```python
new_img_xflip = cv2.flip(img_gray, 0)
```

```python
plt.imshow(new_img_xflip)
```

```python
new_img_yflip = cv2.flip(img_gray, 1)
```

```python
plt.imshow(new_img_yflip)
```

```python
new_img_flip_both = cv2.flip(ar_img, -1)
```

```python
plt.imshow(new_img_flip_both)
```

```python
# the use of fliP ---> generate images for deep learning
```

```python
type(new_img_flip_both)
```

```python
cv2.imwrite("new_flip_both.png", new_img_flip_both)
```

```python
plt.imsave("new_flip_both_plt.png", new_img_flip_both)
```
