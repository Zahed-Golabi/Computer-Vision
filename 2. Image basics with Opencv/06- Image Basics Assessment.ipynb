# Image Basics Assessment

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dog_backpack = cv2.imread("../imgs/dog_backpack.jpg")
```

```python
dog_backpack = cv2.cvtColor(dog_backpack, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dog_backpack)
```

```python
flip_dog_backpack = cv2.flip(dog_backpack, 0)
```

```python
plt.imshow(flip_dog_backpack)
```

```python
plt.imshow(dog_backpack)
```

```python
rectangle_dog_backpack = cv2.rectangle(dog_backpack, pt1=(200,400), pt2=(590,740), color=(255,0,0), thickness=5)
```

```python
plt.imshow(dog_backpack)
```

```python
# triangel vertices
pt1 = (600,800)
pt2 = (500,400)
pt3 = (200,810)
cv2.line(dog_backpack, pt1,pt2, (255,0,0), thickness=10)
cv2.line(dog_backpack, pt1,pt3, (255,0,0), thickness=10)
cv2.line(dog_backpack, pt2,pt3, (255,0,0), thickness=10)

plt.imshow(dog_backpack)
```

```python
triangle_cnt = np.array( [pt1, pt2, pt3] )

cv2.drawContours(dog_backpack, [triangle_cnt], 0, (255,0,0), -1)
plt.imshow(dog_backpack)
```
