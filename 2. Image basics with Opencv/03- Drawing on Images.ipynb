# Drawing on Images

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.uint8)
```

```python
plt.imshow(blank_img)
```

```python
blank_img = cv2.rectangle(blank_img, pt1=(318,0), pt2=(510,150), color=(255,0,0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
blank_img = cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=8)
```

```python
plt.imshow(blank_img)
```

```python
blank_img = cv2.circle(blank_img, center=(150,100), radius=80, color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
# fill the shape thickness=-1
blank_img = cv2.circle(blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
```

```python
plt.imshow(blank_img)
```

```python
blank_img = cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(102,255,255), thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
blank_img = cv2.line(blank_img, pt1=(0,250), pt2=(512,250), color=(39,240, 189), thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
plt.imsave("blank_image.jpg", blank_img)
```

```python
text_img = np.zeros(shape=(512,512,3), dtype=np.uint8)
```

```python
plt.imshow(text_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
text_img = cv2.putText(text_img, text="Hello", org=(10,500), fontFace=font,
           fontScale=4, color=(255,255,255), thickness=3, lineType=cv2.LINE_AA)
```

```python
plt.imshow(text_img)
```

```python
polygon_img = np.zeros(shape=(512,512,3), dtype=np.int32)
```

```python
plt.imshow(polygon_img)
```

```python
vertices = np.array([
    [100,300],
    [200,200],
    [400,300],
    [200,400]
], dtype=np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
pts.shape
```

```python
polygon_img = cv2.polylines(polygon_img, [pts], isClosed=True,
                           color=(255,0,0), thickness=5)
```

```python
plt.imshow(polygon_img)
```
