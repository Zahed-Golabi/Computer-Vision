# Image and Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open("../imgs/flower1.jpg")
```

```python
pic
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
plt.imshow(pic)
```

```python
# Red channe
pic_red = pic_arr[:, :, 0]
```

```python
pic_red.shape
```

```python
plt.imshow(pic_red, cmap="gray")
```

```python
# Green channel
pic_green = pic_arr[:, :, 1]
```

```python
plt.imshow(pic_green, cmap="gray")
```

```python
# Blue channel
pic_blue = pic_arr[:, :, 2]
```

```python
plt.imshow(pic_blue, cmap="gray")
```

```python
full_red = pic_arr.copy()
```

```python
full_red[:, :, 0] = 255
full_red[:, :, 1] = 0
full_red[:, :, 2] = 0
```

```python
plt.imshow(full_red)
```
