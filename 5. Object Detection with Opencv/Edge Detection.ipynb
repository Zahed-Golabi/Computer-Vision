# Edge Detection

- Canny edge detector
 + Canny algorithm: developed in 1986 by John Canny
 + Apply Gaussian filter to smooth the image in order to remove the noise
 + The Canny algorithm also requires a user to decide on low and high threshold values

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imgs/sammy_face.jpg")
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# Lower Threshold to either 0 or 70% of the median_val 
lower = int(max(0, 0.7 * med_val))
# upper Threshold to either 130% of the median_val or the max 255
upper = int(min(255, 1.3 * med_val ))
```

```python
edges = cv2.Canny(img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```

```python
upper
```

```python
# kernel 5*5
blurred_img = cv2.blur(img, ksize=(5,5))
```

```python
edges = cv2.Canny(blurred_img, threshold1=lower, threshold2=upper)
```

```python
plt.imshow(edges)
```

```python
# kernel 7*7
blurred_img = cv2.blur(img, ksize=(7,7))
edges = cv2.Canny(blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```
